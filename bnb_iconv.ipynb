Bnb interval method with convexity

```python
from sympy import *
import interval as ival
import intconv as iconv
import matplotlib.pyplot as plt
import numpy as np
import math
from problems import *
```

```python
iconv_dict = {"cos" : iconv.cos, "sin" : iconv.sin, "exp" : iconv.exp, "log" : iconv.log}
```

```python
problem = problems['xsin5']
print(problem)
```

```python
main_int = iconv.IntConv(problem['ab'])
f = lambdify(x, problem['f'])
fi = lambdify(x, problem['f'], iconv_dict)
print(main_int)
print(fi(main_int))
```

```python
def split(s):
    a = s.x[0]
    b = s.x[1]
    c = s.mid()
    sl = iconv.IntConv([a, c])
    sr = iconv.IntConv([c, b])
    return [sl, sr]
```

```python
subs = [main_int]
eps = problem["eps"]
fr = None
xr = None
i = 0
maxi = 1e4
while len(subs) > 0:
    s = subs.pop()
    print(s, fr)
    c = s.mid()
    fc = f(c)
    if xr == None:
        xr = c
        fr = fc
    elif fc < fr:
        xr = c
        fr = fc
    frange = fi(s)
    print(frange)
    print(len(subs))
    if frange.x[0] < fr - eps:
        subs.extend(split(s))
    i = i + 1
    if i > maxi:
        print("Limit ", maxi, " iteration exceeded\n")
        break
    
```

```python
print(i, xr, fr)
```

```python
def vizual(flist, a, b):
    colors = ['r-', 'b-', 'g-']
    step = (b-a)/100.
    ta = np.arange(a, b + step, step)
    ln = len(ta)
    fta = np.arange(ln).astype(float)
    j = 0
    for f in flist:
        for i in range(ln):
            fta[i] = f(ta[i])
        plt.plot(ta, fta, colors[j])
        j = j + 1
    plt.show()
```

```python
vizual([f], problem['ab'][0], problem['ab'][1])
```
